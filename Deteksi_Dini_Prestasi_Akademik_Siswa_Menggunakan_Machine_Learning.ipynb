## ABSTRAK

Prestasi akademik siswa merupakan indikator penting dalam mengevaluasi efektivitas proses pembelajaran di sekolah menengah. Namun, penilaian akademik sering kali bersifat reaktif karena dilakukan setelah nilai akhir diperoleh, sehingga peluang intervensi dini menjadi terbatas. Oleh karena itu, penelitian ini bertujuan untuk mengembangkan model machine learning yang mampu memprediksi nilai akhir siswa (G3) serta mengidentifikasi faktor-faktor yang paling berpengaruh terhadap capaian akademik siswa.

Dataset yang digunakan berasal dari dua sekolah menengah di Portugal dan mencakup informasi akademik, karakteristik keluarga, lingkungan, serta perilaku siswa. Proses pemodelan dilakukan menggunakan pendekatan regresi non-linear dengan beberapa algoritma ensemble tree, yaitu Random Forest, Gradient Boosting, dan ExtraTrees Regressor. Untuk meningkatkan performa dan stabilitas model, dilakukan hyperparameter tuning menggunakan RandomizedSearchCV dengan skema validasi silang. Evaluasi model dilakukan menggunakan metrik Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), dan R-Squared (R²).

Hasil pemodelan menunjukkan bahwa ExtraTrees Regressor merupakan model terbaik dengan nilai R² sebesar 0,86 pada data uji, yang berarti model mampu menjelaskan sekitar 86% variasi nilai akhir siswa, serta menghasilkan MAE sekitar ±0,70, yang menunjukkan tingkat kesalahan prediksi yang relatif rendah. Analisis feature importance mengungkapkan bahwa nilai akademik sebelumnya (G1 dan G2) merupakan faktor paling dominan dalam menentukan nilai akhir siswa, diikuti oleh riwayat kegagalan akademik dan tingkat kehadiran. Faktor keluarga, lingkungan, dan perilaku siswa memberikan pengaruh yang lebih kecil namun tetap konsisten sebagai faktor pendukung.

Berdasarkan hasil tersebut, model yang dikembangkan dinilai layak digunakan sebagai sistem pendukung keputusan untuk deteksi dini siswa berisiko rendah prestasi akademik. Model ini dapat membantu sekolah dan pendidik dalam merancang intervensi akademik yang lebih tepat sasaran sejak awal semester, sehingga diharapkan dapat meningkatkan capaian akademik siswa secara keseluruhan.

### KENAPA MAE, RMSE, R2?

Dalam konteks prediksi nilai akhir siswa (G3), tujuan utama model regresi bukan hanya akurasi matematis, tetapi kegunaan praktis untuk pengambilan keputusan akademik. Oleh karena itu, metrik evaluasi yang dipilih harus:

mudah dipahami oleh non-teknis (guru, konselor, manajemen sekolah),

mencerminkan kesalahan prediksi secara nyata, dan

relevan terhadap risiko keputusan pendidikan.

Berdasarkan pertimbangan tersebut, metrik yang digunakan adalah MAE, RMSE, dan R-Squared. Berikut penjelasan lebih detail terkait penggunaan MAE, RMSE, dan R-Squared:

1. Mean Absolute Error (MAE)
      
      Alasan Industri

      MAE mengukur rata-rata selisih absolut antara nilai prediksi dan nilai aktual siswa.

      Dalam konteks pendidikan:

      - MAE menjawab pertanyaan praktis:
        
        “Secara rata-rata, model meleset berapa poin nilai dari nilai sebenarnya?”

      - Misalnya MAE = 1.8 berarti:

        prediksi nilai akhir siswa rata-rata meleset sekitar 1–2 poin.

      Ini sangat penting bagi industri pendidikan karena:

      Selisih 1–2 poin masih dapat diterima untuk keputusan intervensi akademik.

      MAE tidak sensitif terhadap outlier, sehingga stabil ketika terdapat siswa dengan nilai ekstrem (sangat tinggi atau sangat rendah).

      Kenapa MAE penting secara industri

      - Guru dan konselor lebih mudah memahami kesalahan absolut daripada kuadrat error.

      - Cocok untuk early warning system, di mana tujuan utama adalah perkiraan kasar tapi stabil, bukan penalti ekstrem.

2. Root Mean Squared Error (RMSE)

      Alasan Industri

      RMSE memberikan penalti lebih besar pada kesalahan prediksi yang besar, karena error dikuadratkan sebelum dirata-rata.

      Dalam konteks industri pendidikan:

      - RMSE digunakan untuk mengidentifikasi risiko kesalahan besar, misalnya:

        - siswa diprediksi lulus, tetapi ternyata gagal jauh,

        - atau siswa diprediksi sangat rendah, padahal performanya baik.

      RMSE menjawab pertanyaan:

      “Seberapa parah kesalahan terburuk yang cenderung dilakukan model?”

      Kenapa RMSE tetap dibutuhkan

      - Kesalahan besar dalam prediksi akademik memiliki konsekuensi serius:

        - intervensi salah sasaran,

        - siswa tidak mendapat dukungan yang seharusnya,

        - atau salah alokasi sumber daya sekolah.

      - RMSE membantu mengontrol risiko ekstrem, yang tidak tertangkap oleh MAE.

      Hubungan MAE & RMSE (Sudut Industri)

      - MAE itu akurasi rata-rata (operasional)

      - RMSE itu risiko kesalahan besar (risk control)

      Industri hampir selalu menggunakan keduanya bersamaan, bukan salah satu.

3. R-squared (R²)

      Alasan Industri

      R² mengukur seberapa besar variasi nilai akhir siswa (G3) yang dapat dijelaskan oleh model.

      Dalam konteks ini:

      - R² menjawab pertanyaan strategis:

      “Seberapa besar performa akademik siswa bisa dijelaskan oleh data yang kita miliki?”

      Ini penting bagi manajemen sekolah karena:

      - R² tinggi maka faktor dalam data cukup representatif

      - R² rendah maka masih banyak faktor eksternal (psikologis, lingkungan luar sekolah) yang belum tertangkap

      Kenapa R² tetap dipakai (meski bukan metrik utama)

      - Digunakan sebagai indikator kualitas model secara global, bukan keputusan operasional.

      - Membantu komunikasi dengan stakeholder tentang batas kemampuan model.

**Kenapa Tidak Menggunakan Metrik Lain?**

| Metrik | Alasan Tidak Digunakan |
|--------|------------------------|
|MSE | Tidak intuitif (satuan kuadrat nilai) |
|MAPE | Tidak stabil untuk nilai kecil / nol |
| MSLE | Tidak relevan karena skala nilai terbatas |
| Adjusted R-Squared | Tidak krusial untuk predictive task |
| Huber Loss | Lebih cocok untuk training, bukan reporting |

### UJI NORMALITAS

```python
result = {"Test":["Nilai KS-Test","Hasil KS-Test","Nilai Shapiro","Hasil Shapiro","Nilai NormalTest","Hasil NormalTest","Nilai Lilliefors","Hasil Lilliefors"]}

for kolom in dataangka.columns:
  statkstest,pvaluekstest = kstest(dataangka[kolom],"norm")
  statshapiro,pvalueshapiro = shapiro(dataangka[kolom])
  statnormal,pvaluenormal = normaltest(dataangka[kolom])
  statlilliefors,pvaluelilliefors = lilliefors(dataangka[kolom])

  if pvaluekstest > 0.05:
    hasilkstest = "Distribusi Normal"
  else:
    hasilkstest = "Distribusi Tidak Normal"
  if pvalueshapiro > 0.05:
    hasilshapiro = "Distribusi Normal"
  else:
    hasilshapiro = "Distribusi Tidak Normal"
  if pvaluenormal > 0.05:
    hasilnormal = "Distribusi Normal"
  else:
    hasilnormal = "Distribusi Tidak Normal"
  if pvaluelilliefors > 0.05:
    hasillilliefors = "Distribusi Normal"
  else:
    hasillilliefors = "Distribusi Tidak Normal"

  result[kolom] = [pvaluekstest,hasilkstest,pvalueshapiro,hasilshapiro,pvaluenormal,hasilnormal,pvaluelilliefors,hasillilliefors]

result = pd.DataFrame(result).set_index("Test").T
result
```

Pada bagian uji normalitas, data di uji dengan metode KS-Test, Shapiro Test, Normal Test, Lilliefors Test untuk membuktikan secara distribusi data. Dari hasil pengujian terhadap setiap kolom dengan setiap metode didapatkan informasi mayoritas bahwa data numerik berdistribusi tidak normal, tetapi terdapat 1 kolom yaitu G1 dengan 1 metode yaitu normal test menunjukkan informasi hasil uji normalitas adalah distribusi normal. ## CEK MULTIKOLINEARITAS

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
import statsmodels.api as sm
```

```python
# 1) Ambil X (pastikan target tidak ikut)
Cek = dataangkanew.drop(columns="G3")

# 2) (Opsional tapi aman) boolean tetap dipertahankan tapi diubah ke 0/1
bool_cols = Cek.select_dtypes(include=["bool"]).columns
Cek[bool_cols] = Cek[bool_cols].astype(int)

# 3) Pastikan semua kolom numerik (kalau ada yang nyasar jadi object, akan jadi NaN)
Cek = Cek.apply(pd.to_numeric, errors="coerce")

# 4) Bersihkan inf / NaN
Cek = Cek.replace([np.inf, -np.inf], np.nan).dropna()

# 5) Tambah intercept SATU KALI
Ceksm = sm.add_constant(Cek, has_constant="add")

# 6) Kunci perbaikan: paksa ke NumPy float (hindari .values yang bisa jadi object karena Int64)
X = Ceksm.to_numpy(dtype=float)

# 7) Hitung VIF
vif_df = pd.DataFrame({
    "Variable": Ceksm.columns,
    "VIF": [variance_inflation_factor(X, i) for i in range(X.shape[1])]
})

vif_df
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, cross_validate
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.dummy import DummyRegressor
from sklearn.ensemble import (
    RandomForestRegressor,
    ExtraTreesRegressor,
    GradientBoostingRegressor,
    HistGradientBoostingRegressor
)

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scoring = {
    "mae": "neg_mean_absolute_error",
    "rmse": "neg_root_mean_squared_error",
    "r2": "r2"
}
```

```python
models = [
    # Bagging Trees
    ("RandomForest", RandomForestRegressor()),
    ("ExtraTrees", ExtraTreesRegressor()),

    # Boosting (sklearn)
    ("GradientBoosting", GradientBoostingRegressor()),
    ("HistGradientBoosting", HistGradientBoostingRegressor()),

    # Boosting modern
    ("XGBoost", XGBRegressor()),
    ("LightGBM", LGBMRegressor()),

    # CatBoost (pakai transformer = fair comparison)
    ("CatBoost", CatBoostRegressor(verbose=0))
]
```

```python
def evaluate_models(
    models,
    transformer,
    X_train, y_train,
    X_test, y_test,
    cv,
    scoring
):
    rows = []

    for name, model in models:
        pipe = Pipeline([
            ("transformer", transformer),
            ("model", model)
        ])

        cv_result = cross_validate(
            pipe,
            X_train,
            y_train,
            cv=cv,
            scoring=scoring,
            return_train_score=False,
            n_jobs=-1
        )

        cv_mae_mean  = -cv_result["test_mae"].mean()
        cv_mae_std   =  cv_result["test_mae"].std()

        cv_rmse_mean = -cv_result["test_rmse"].mean()
        cv_rmse_std  =  cv_result["test_rmse"].std()

        cv_r2_mean   =  cv_result["test_r2"].mean()
        cv_r2_std    =  cv_result["test_r2"].std()

        pipe.fit(X_train, y_train)
        y_pred = pipe.predict(X_test)

        rows.append({
            "Model": name,

            "CV_MAE_mean": cv_mae_mean,
            "CV_MAE_std": cv_mae_std,

            "CV_RMSE_mean": cv_rmse_mean,
            "CV_RMSE_std": cv_rmse_std,

            "CV_R2_mean": cv_r2_mean,
            "CV_R2_std": cv_r2_std,

            "Test_MAE": mean_absolute_error(y_test, y_pred),
            "Test_RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
            "Test_R2": r2_score(y_test, y_pred)
        })

    return (
        pd.DataFrame(rows)
        .sort_values(by="CV_MAE_mean", ascending=True)
        .reset_index(drop=True)
    )
```

```python
results = evaluate_models(
    models=models,
    transformer=transformer,
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    cv=kf,
    scoring=scoring
)

results
```

## HYPERPARAMETER TUNING

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, RandomizedSearchCV, cross_validate
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor, GradientBoostingRegressor

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scoring_multi = {
    "mae": "neg_mean_absolute_error",
    "rmse": "neg_root_mean_squared_error",
    "r2": "r2"
}

def tune_and_compare(name, base_model, param_distributions, transformer,
                     X_train, y_train, X_test, y_test,
                     cv, n_iter=50, random_state=42):
    pipe = Pipeline([
        ("transformer", transformer),
        ("model", base_model)
    ])

    search = RandomizedSearchCV(
        estimator=pipe,
        param_distributions=param_distributions,
        n_iter=n_iter,
        scoring="neg_mean_absolute_error",
        refit=True,
        cv=cv,
        n_jobs=-1,
        random_state=random_state,
        verbose=1
    )
    search.fit(X_train, y_train)

    best_pipe = search.best_estimator_

    cv_res = cross_validate(
        best_pipe,
        X_train, y_train,
        cv=cv,
        scoring=scoring_multi,
        return_train_score=True,
        n_jobs=-1
    )

    train_mae_mean  = -cv_res["train_mae"].mean()
    train_mae_std   =  cv_res["train_mae"].std()
    train_rmse_mean = -cv_res["train_rmse"].mean()
    train_rmse_std  =  cv_res["train_rmse"].std()
    train_r2_mean   =  cv_res["train_r2"].mean()
    train_r2_std    =  cv_res["train_r2"].std()

    cv_mae_mean  = -cv_res["test_mae"].mean()
    cv_mae_std   =  cv_res["test_mae"].std()
    cv_rmse_mean = -cv_res["test_rmse"].mean()
    cv_rmse_std  =  cv_res["test_rmse"].std()
    cv_r2_mean   =  cv_res["test_r2"].mean()
    cv_r2_std    =  cv_res["test_r2"].std()

    best_pipe.fit(X_train, y_train)
    pred = best_pipe.predict(X_test)

    test_mae  = mean_absolute_error(y_test, pred)
    test_rmse = np.sqrt(mean_squared_error(y_test, pred))
    test_r2   = r2_score(y_test, pred)

    row = {
        "Model": name,

        "Train_MAE_mean": train_mae_mean,
        "Train_MAE_std":  train_mae_std,
        "Train_RMSE_mean": train_rmse_mean,
        "Train_RMSE_std":  train_rmse_std,
        "Train_R2_mean":  train_r2_mean,
        "Train_R2_std":   train_r2_std,

        "CV_MAE_mean": cv_mae_mean,
        "CV_MAE_std":  cv_mae_std,
        "CV_RMSE_mean": cv_rmse_mean,
        "CV_RMSE_std":  cv_rmse_std,
        "CV_R2_mean":  cv_r2_mean,
        "CV_R2_std":   cv_r2_std,

        "Test_MAE": test_mae,
        "Test_RMSE": test_rmse,
        "Test_R2": test_r2,

        "Best_Params": search.best_params_
    }

    return row, search
```

### PARAMETER YANG DIGUNAKAN DAN ALASANNYA

1) n_estimators (jumlah trees)

  - Mengontrol stabilitas prediksi (lebih banyak tree maka variance turun).

  - Biasanya meningkatkan performa sampai titik jenuh.

  - Dituning karena dampaknya besar dan relatif “aman”.

2) max_depth

  - Mengontrol kompleksitas pohon.

  - Terlalu dalam maka overfitting, terlalu dangkal maka underfitting.

  - Dituning karena dataset Anda banyak pola non-linear dan interaksi.

3) min_samples_split

  - Mengatur kapan node boleh di-split.

  - Nilai lebih besar maka model lebih “halus” dan lebih tahan noise.

  - Penting untuk mengontrol overfitting.

4) min_samples_leaf

  - Minimal jumlah sampel di daun.

  - Nilai lebih besar maka prediksi lebih stabil (terutama bila data ada outlier/noise).

  - Ini salah satu regularisasi paling efektif di tree ensemble.

5) max_features

  - Berapa banyak fitur yang dipertimbangkan saat split.

  - Mengontrol trade-off bias–variance dan “keragaman” antar tree.

  - Sangat relevan pada ensemble (RandomForest/ExtraTrees).

6) bootstrap

  - Apakah sampling data untuk tiap tree memakai bootstrap.

  - False (default ExtraTrees) sering bagus, tapi True kadang membantu generalisasi.

  - Layak dituning sebagai opsi regularisasi tambahan.

```python
param_et = {
    "model__n_estimators": [300, 600, 900, 1200],
    "model__max_depth": [None, 5, 10, 15, 20, 30],
    "model__min_samples_split": [2, 5, 10, 20],
    "model__min_samples_leaf": [1, 2, 4, 8, 12],
    "model__max_features": ["sqrt", "log2", None, 0.5, 0.8],
    "model__bootstrap": [False, True]
}
```

```python
rows = []

row_et, search_et = tune_and_compare(
    "ExtraTrees_tuned",
    ExtraTreesRegressor(random_state=42, n_jobs=-1),
    param_et,
    transformer,
    X_train, y_train, X_test, y_test,
    cv=kf,
    n_iter=50
)
rows.append(row_et)

tuned_results = pd.DataFrame(rows).sort_values("CV_MAE_mean").reset_index(drop=True)
tuned_results
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import ExtraTreesRegressor

# parameter terbaik dari tuning kamu
best_params_et = {
    "n_estimators": 300,
    "min_samples_split": 5,
    "min_samples_leaf": 4,
    "max_features": None,
    "max_depth": 5,
    "bootstrap": False
}

final_model = Pipeline([
    ("transformer", transformer),
    ("model", ExtraTreesRegressor(random_state=42, n_jobs=-1, **best_params_et))
])

final_model.fit(X_train, y_train)
```
